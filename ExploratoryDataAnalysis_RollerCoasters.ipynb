##Step 5: Ask Question about the Data

1) What are the locations with the fastest rollercoasters (min of 10 coasters)?


```python
# df["Location"].value_counts() #we dont want Other so
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10'). sort_values('mean')['mean'] \
    .plot(kind='barh', title='Average Speed by Location')

ax.set_xlabel('Average Speed by Location')
plt.show()
```

2) What is the average, minimum, and maximum speed (mph) and height (ft) of the rollercoasters?

```python
df[['Speed_mph', 'Height_ft']].agg(['mean', 'min', 'max'])
```

3) How many rollercoasters have inversions? What is the average number of inversions?

```python
df.query("Inversions > 0")['Inversions'].count()
```

```python
df.query("Inversions > 0")['Inversions'].agg(['mean'])
```

4) What is the distribution of rollercoaster heights? (Use a histogram or KDE plot)

10)	Is there a correlation between speed and height of the rollercoasters?

11) How does the number of inversions correlate with g-force?

```python
df_2 = df[['Inversions', 'Gforce']]
ax = sns.scatterplot(x='Inversions',
                y='Gforce',
                data=df_2)
plt.title('Correlation between Inversions and Gforce')
ax.set_xlabel('Number of Inversions')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Inversions'], df['Gforce'], alpha=0.6)
plt.title('Scatter Plot of Inversions vs G-force')
plt.xlabel('Number of Inversions')
plt.ylabel('G-force')
plt.grid(True)
plt.show()
```
